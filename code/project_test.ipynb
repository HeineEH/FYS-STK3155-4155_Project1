### Part a: Ordinary Least Square (OLS) for the Runge function

```python
# Imports
import numpy as np
```

```python
# Defining the Runge function
def Runge(x):
    return 1/(1+25*x**2)
```

```python
n = 100
x = np.random.uniform(-1, 1, n)
y = Runge(x) + 0.05*np.random.normal(0, 1, n)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, Runge(x)) 
plt.scatter(x, y)
plt.show()
```

```python
def OLS_parameters(X, y):
    X_transpose = np.transpose(X)
    return np.linalg.pinv(X_transpose @ X) @ X_transpose @ y
```

```python
def polynomial_features(x, p):
    n = len(x)
    X = np.zeros((n, p+1))
    for i in range(p+1): 
        X[:, i] = x**i
    return X

POLYNOMIAL_DEGREE = 15
X = polynomial_features(x, POLYNOMIAL_DEGREE)
```

```python
theta = OLS_parameters(X, y)
```

```python
def Ridge_parameters(X, y, lambda_reg = 0.1):
    pred = X.shape[1]
    I = np.eye(pred) # size: (p,p)

    return np.linalg.inv((X.T @ X)+I*lambda_reg) @ X.T @ y

```
